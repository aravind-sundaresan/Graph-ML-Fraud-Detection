```python
import numpy as np
import pandas as pd

banksim_df = pd.read_csv("../data/bs140513_032310.csv")
banksim_df.info()
```

```python
banksim_df = banksim_df.drop(['step', 'merchant', 'age', 'gender', 'zipcodeOri', 'amount', 'zipMerchant'], axis=1)
banksim_df.info()
```

```python
count_per_transaction_type = banksim_df.groupby(['customer', 'fraud']).size().unstack().add_prefix("fraud_")
count_per_transaction_type.head()
```

```python
count_per_category = banksim_df.groupby(['customer', 'category']).size().unstack()
count_per_category.head()
```

```python
transaction_count_per_group = pd.merge(count_per_transaction_type, count_per_category, left_index=True, right_index=True)
transaction_count_per_group.head()
```

```python
transaction_count_per_group = transaction_count_per_group.drop(['\'es_barsandrestaurants\'', '\'es_contents\'', '\'es_hotelservices\'', '\'es_hyper\'', '\'es_otherservices\'', '\'es_sportsandtoys\'', '\'es_tech\'', '\'es_travel\'', '\'es_wellnessandbeauty\''], axis=1)
transaction_count_per_group.info()
```

```python
# Renaming the columns
transaction_count_per_group.columns = ["legitimate_count", "fraudulent_count", "Fashion", "Food", "Health", "Home", "Leisure", "Transportation"]
transaction_count_per_group['customer_id'] = transaction_count_per_group.index
transaction_count_per_group.info()
```

```python
# Reordering the columns
transaction_count_per_group = transaction_count_per_group[["customer_id", "legitimate_count", "fraudulent_count", "Transportation", "Food", "Fashion", "Leisure", "Health", "Home"]]
transaction_count_per_group.head()
```

```python
transaction_count_per_group = transaction_count_per_group.fillna(0)
transaction_count_per_group.head()
```

```python
transaction_count_per_group.to_csv("sample.csv", index=False)
```
